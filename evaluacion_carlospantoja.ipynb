```python
# 8. Ahora grafica los datos de "Close" usando la biblioteca matplotlib en Python, 
# 9. Agrega title, marker, linestyle y color para mejorar la visualizacion

close = data['Close']
ax = close.plot(title='Facebook', linestyle='-', color='b')
ax.set_xlabel('Years')
ax.set_ylabel('Close')
ax.grid() #opcional
plt.show()
```

* El grafico de linea nos muestra la evolución de los precios de cierre de las acciones de Facebook, durante la gestion Enero 2020 a agosto 2021, observamos que desde enero hasta marzo los precios sufrieron una caida, sin embargo a partir de mayor del 2020 tuvieron un comportamiento creciente, este incremento se mantuvo moderado hasta mayo del 2021, luego los precios tuvieron un incremento mucho mas significativo. La serie de datos de las acciones muestra un comportamiento tendencial y estocastico por lo que al no ser estacionario es dificil predecir el comportamiento de este.
